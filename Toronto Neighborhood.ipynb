ASSIGNING WIKIPEDIA URL IN A VARIABLE url

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
```

UTILIZING BEAUTIGULSOUP LIBRARY FOR WEB SCRAPING

```python
page = urllib.request.urlopen(url)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(page, "lxml")
```

```python
print(soup.prettify())
```

ASSINGING HTML TABLE TO VARABLE table

```python
table = soup.find('table', class_= 'wikitable sortable')
table
```

CREATING 3 EMPTY LISTS AND APPENDING THE SCRAPED DATA TO THE LISTS, TR AND TD ARE THE TAGS BETWEEN ROWS AND AND INDIVIDUAL VALUES IN THE SCRAPED DATA

```python
a= []
b=[]
c=[]
for row in table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==3:
        a.append(cells[0].find(text=True))
        b.append(cells[1].find(text=True))
        c.append(cells[2].find(text=True))
```

REMOVING NEW LINE CHARACTER FROM ALL THE VALUES

```python
a1 =[]
b1 =[]
c1 =[]
import re
for x in a:
    a1.append(re.sub('\n', '', x))
for x in b:
    b1.append(re.sub('\n', '', x))
for x in c:    
    c1.append(re.sub('\n', '', x))    

df=pd.DataFrame(a1,columns=['postalcode'])
df['borough']=b1
df['neighborhood']=c1
df.head(2)
```

DROPPING ROWS WHERE BOROUGH IS 'NOT ASSIGNED'

```python
df.drop(df.loc[df['borough']=='Not assigned'].index, inplace=True)
```

```python
import folium
# create map of Manhattan using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, label in zip(df1['Latitude'], df1['Longitude'], df1['neighborhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
